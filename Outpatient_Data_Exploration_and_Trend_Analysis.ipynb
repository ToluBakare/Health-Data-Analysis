# Trend Series

To start we have to change the 'Visit_Date' column to date format and set it as index

```python
df_visit['Visit_Date'][0]
```

```python
type(df_visit['Visit_Date'][0])
```

```python
df_visit['Visit_Date'] = pd.to_datetime(df_visit['Visit_Date'])
```

```python
type(df_visit['Visit_Date'][0])
```

```python
df_visit.set_index('Visit_Date', inplace=True)
```

```python
df_visit
```

Data is grouped by year. We have 5 years, making five distinct groups

```python
df_resampled_year = df_visit.resample('YE').count()
```

```python
df_resampled_year
```

Plotting the graph for the trend analysis per year

```python

plt.figure (figsize=(10,5))
plt.plot(df_resampled_year.index, df_resampled_year['Patient_ID'])
plt.title('Trend analysis of Yearly Visits')
plt.xlabel('Date')
plt.ylabel('Number of Visit')
plt.grid(True)

plt.show()
```

*2021 had a sharp decline*

Plotting by month to have a better insight

```python
df_resampled_month = df_visit.resample('ME').count()
```

```python
plt.figure (figsize=(10,5))
plt.plot(df_resampled_month.index, df_resampled_month['Patient_ID'])
plt.title('Trend analysis of Monthly Visits')
plt.xlabel('Date')
plt.ylabel('Number of Visit')
plt.grid(True)

plt.show()
```

Adding ticks for better view

```python
monthly_ticks =pd.date_range(start= df_resampled_month.index.min(), end= df_resampled_month.index.max(), freq='ME')
monthly_labels = [date.strftime('%b%Y') for date in monthly_ticks]
plt.figure(figsize=(15,5))
plt.plot(df_resampled_month.index, df_resampled_month['Patient_ID'])
plt.title('Trend analysis of Monthly Visits')
plt.xlabel('Date')
plt.xticks(monthly_ticks, monthly_labels, rotation=90)
plt.ylabel('Number of Visit')
plt.grid(True)

plt.show()
```

Now, we see February seem to have deep declines. We will dig deeper to get more insights on Februarys

```python
from time import strftime
target_month = 2
df_target_month= df_resampled_month[df_resampled_month.index.month == target_month]

X_labels= [date.strftime('%b%Y') for date in df_target_month.index]
plt.plot(df_target_month.index, df_target_month['Patient_ID'])
plt.title(f"Trend analysis of {df_target_month.index[0].strftime('%b')} Across Five Years")
plt.xticks(df_target_month.index, X_labels)
plt.xlabel('Date')
plt.ylabel('Number of Visits')
plt.grid(True)

plt.show()
```

*What happened in February 2020? Covid?*

Taking a look at Novembers.

```python
from time import strftime
target_month = 11
df_target_month= df_resampled_month[df_resampled_month.index.month == target_month]

X_labels= [date.strftime('%b%Y') for date in df_target_month.index]
plt.plot(df_target_month.index, df_target_month['Patient_ID'])
plt.title(f"Trend analysis of {df_target_month.index[0].strftime('%b')} Across Five Years")
plt.xticks(df_target_month.index, X_labels)
plt.xlabel('Date')
plt.ylabel('Number of Visits')
plt.grid(True)

plt.show()
```
